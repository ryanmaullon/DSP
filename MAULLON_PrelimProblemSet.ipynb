##### <b>Part 1.</b> [150 pts] Solve for the magnitude and angle of the linear transformations given. Provide necessary solutions numerically (hand-solved) and computationally (using a Python program). Also provide the necessary solutions vector plots of the given matrices and their transformations.
$$\mathbf{H} = \frac{1}{\sqrt{2}} \begin{pmatrix} {1} & {1} \\ {1} & {-1} \end{pmatrix} , \mathbf{Y} = \begin{pmatrix} {0} & {-i} \\ {i} & {0} \end{pmatrix} ,\mathbf{O} = \begin{pmatrix} {1} & {0} \\ {1} & {1} \end{pmatrix}


```python
import numpy as np
import matplotlib.pyplot as plt

#Define Matrix H,Y,O
H = (1 / np.sqrt(2)) * np.array([[1, 1], [1, -1]])
Y = np.array([[0, -1j], [1j, 0]])
O = np.array([[1,0],[1,1]])
```

```python
# 1. H⋅Y
# Compute the linear transformation H⋅Y
HY = np.dot(H, Y)
print("Product:\n", HY)

# Find the magnitude of HY
magnitude_HY = np.linalg.norm(np.round(HY,2))
print("Magnitude of HY:", magnitude_HY)

# Find the angle of HY
angle_HY = np.angle(HY, deg=True)
print("Angle of HY:","\n",angle_HY)

# Represent matrices as complex numbers
H_complex = H[0, 0] + H[0, 1] * 1j
Y_complex = Y[0, 0] + Y[0, 1] * 1j
HY_complex = HY[0, 0] + HY[0, 1] * 1j

plt.figure(figsize=(5, 5))
# Plot H as red arrow
plt.quiver(0, 0, H_complex.real, H_complex.imag, angles='xy', scale_units='xy', scale=1, color='r', label='H')

# Plot Y as green arrow
plt.quiver(0, 0, Y_complex.real, Y_complex.imag, angles='xy', scale_units='xy', scale=1, color='g', label='Y')

# Plot HY as blue arrow
plt.quiver(0, 0, HY_complex.real, HY_complex.imag, angles='xy', scale_units='xy', scale=1, color='b', label='H⋅Y')

plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Vector Plot')
plt.legend()
plt.grid(True)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
# 2. Y⋅H
# Compute the linear transformation Y⋅H
YH = np.dot(Y,H)
print("Product:\n",YH)

# Find the magnitude of YH
magnitude_YH = np.linalg.norm(YH)
print("Magnitude of YH:", np.round(magnitude_YH,2))

# Find the angle of YH
angle_YH = np.angle(YH, deg=True)
print("Angle of YH:","\n",angle_YH)

# Represent matrices as complex numbers
H_complex = H[0, 0] + H[0, 1] * 1j
Y_complex = Y[0, 0] + Y[0, 1] * 1j
YH_complex = YH[0, 0] + YH[0, 1] * 1j

plt.figure(figsize=(5, 5))
# Plot H as red arrow
plt.quiver(0, 0, H_complex.real, H_complex.imag, angles='xy', scale_units='xy', scale=1, color='r', label='H')

# Plot Y as green arrow
plt.quiver(0, 0, Y_complex.real, Y_complex.imag, angles='xy', scale_units='xy', scale=1, color='g', label='Y')

# Plot YH as blue arrow
plt.quiver(0, 0, YH_complex.real, YH_complex.imag, angles='xy', scale_units='xy', scale=1, color='b', label='Y⋅H')

plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Vector Plot')
plt.legend()
plt.grid(True)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()
```

##### <b>Part 3.</b> [50 pts] Determine whether the resulting linear transformations are linearly dependent. Provide necessary solutions both numerically using determinants (hand-solved) and computationally (using a Python program).

```python
# NOTE: If the determinant of the matrix is equal to 0, it is a Singular Matrix and Linearly dependent. Otherwise, it is Non Singular Matrix and Linearly independent
#NUMBER 1
A1 = np.array([
    [5,0,0],
    [0,5,0],
    [0,0,5]])
B1 = np.array([
    [1,0,5],
    [2,7,6],
    [6,4,7]])  
result1 = np.dot(A1,B1)
det1 = np.linalg.det(result1)
print(result1)
print("Determinant:",det1)
if det1 == 0:
    print("The linear transformation is Linearly dependent")
else:
    print("The linear transformation is Linearly Independent")
```

```python
#NUMBER 2
A2 = np.array([
    [1,2,6],
    [3,15,4],
    [2,10,3]])
B2 = np.array([
    [5,2,4],
    [6,2,4],
    [0,1,1]])
result2 = np.dot(A2,B2)
det2 = np.linalg.det(result2)
print(result2)
print("Determinant:",det2)
if det2 == 0:
    print("The linear transformation is Linearly dependent")
else:
    print("The linear transformation is Linearly Independent")
```

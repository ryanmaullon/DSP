##### <b>Part 2.</b> [50 pts] Compute for the Magnitude and Phase of Part 1 number 1-9. You should denote the magnitude as M_x and Phase as _x where x is the variable name of the complex number.



```python
import numpy as np

### FOR COMPUTING MAGNITUDE OF THE COMPLEX NUMBER USING ITERATION
# Define the complex numbers
z1 = 1 - 5j
z2 = 3 + 2j
z3 = -7 + 1j
z4 = -1.6j

# Store z1 to z4 complex numbers to array
complex_numbers = [z1, z2, z3, z4]

# Define and calculate complex numbers z5 to w
z5 = z1 * z1.conjugate()
z6 = z2 * (z1 - z3).conjugate()
z7 = z4 / z1
z8 = (z4 / z2) * (z1 + z4)
z9 = (z1 / z3) + ((2 * z4) / 3)
w = np.exp(z4)

# Append z5 to z10 to the complex_numbers list
complex_numbers.extend([z5, z6, z7, z8, z9, w])

# Array for magnitude
magnitudes = []

# Calculate the magnitudes of z1 to w
for z in complex_numbers:
    magnitude = np.abs(z)
    magnitudes.append(magnitude)

# Print the magnitudes
for i, magnitude in enumerate(magnitudes):
    variable_name = f'z{i + 1}' if i < 9 else 'w'
    print(f'Magnitude of {variable_name}: {magnitude}')
```

```python
import numpy as np

### FOR COMPUTING PHASE OF THE COMPLEX NUMBER USING ITERATION
# Define the complex numbers
z1 = 1 - 5j
z2 = 3 + 2j
z3 = -7 + 1j
z4 = -1.6j

# Store z1 to z4 complex numbers to array
complex_numbers = [z1, z2, z3, z4]

# Define and calculate complex numbers z5 to w
z5 = z1 * z1.conjugate()
z6 = z2 * (z1 - z3).conjugate()
z7 = z4 / z1
z8 = (z4 / z2) * (z1 + z4)
z9 = (z1 / z3) + ((2 * z4) / 3)
w = np.exp(z4)

# Append z5 to z10 to the complex_numbers list
complex_numbers.extend([z5, z6, z7, z8, z9, w])

# Array for angles
angles = []

# Calculate the magnitudes of z1 to w
for z in complex_numbers:
    angle = np.angle(z,deg=True)
    angles.append(angle)

# Print the magnitudes
for i, angles in enumerate(angles):
    variable_name = f'z{i + 1}' if i < 9 else 'w'
    print(f'Angle of {variable_name}: {angles}')
```
